## Manipulating data

```python
car_sales["Make"] = car_sales["Make"].str.lower()
```

```python
car_sales
```

```python
car_sales_missing = pd.read_csv("../data/car-sales-missing-data.csv")
car_sales_missing
```

```python
car_sales_missing["Odometer"].fillna(0, inplace=True)
car_sales_missing
```

```python
car_sales_missing.dropna(inplace=True)
car_sales_missing
```

```python
# Column from series
seats_column = pd.Series([5, 5, 5, 3, 5])

# New column
car_sales["Seats"] = seats_column
car_sales
```

```python
car_sales["Seats"].fillna(0, inplace=True)
car_sales
```

```python
# Column from Python list
fuel_economy = [8.8, 7.2, 9.6, 11.3, 6.6, 7.8, 8.6, 5.5, 9.7, 6.9]
car_sales["Fuel per 100KM"] = fuel_economy # needs to be the same lenght as entries
car_sales
```

```python
car_sales["Total fuel used (L)"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
car_sales
```

```python
# Create a column from a single value
car_sales["Number of wheels"] = 4
car_sales
```

```python
car_sales = car_sales.drop("Total fuel used (L)", axis=1)
car_sales
```

```python
# shuffle the complete data set
car_sales_shuffled = car_sales.sample(frac=1)
car_sales_shuffled
```

```python
# shuffle only 20% of the data
car_sales_shuffled_new = car_sales_shuffled.sample(frac=0.2)
car_sales_shuffled_new
```

```python
car_sales_shuffled.reset_index(drop=True,inplace=True)
car_sales_shuffled
```

```python
# convert data within 1 column
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x: x / 1.6)
car_sales.rename(columns={"Odometer (KM)": "Odometer (M)"}, inplace=True)
car_sales
```
